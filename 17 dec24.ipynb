```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("D:\data enegenering\datasets\datasets\score.csv")
df.head()
```

```python
x = df['marks']
y = df['rank']
z = df.admit
```

```python
x1 = len(x[x>950])
```

```python
x2 = len(x[(x>850) & (x<=950)])
x3 = len(x[x<=850])
print(x1, x2, x3)
```

```python
slices = [x1, x2, x3]
cols = ['yellow', 'red', 'green']
lbs = ['marks>950', 'marks 850-950', 'marks<950']
```

```python
plt.pie(slices, colors=cols, labels=lbs)
plt.show()
```

```python
plt.pie(slices, colors=cols, labels=lbs, startangle=90, shadow=True)
plt.show()
```

```python
plt.pie(slices, colors = cols, labels=lbls, startangle=90, shadow=True, autoct = '%.2f%%)
plt.show()
```
